```python
import random
from gtts import gTTS
import os
from IPython.display import Audio

# Predefined user credentials for authentication
credentials = {
    "admin": {"username": "admin", "password": "admin123"},
    "hr": {"username": "hr", "password": "hr123"},
    "user": {"username": "user", "password": "user123"},
}

# Authentication Function
def authenticate(role):
    """
    Authenticates a user based on their role.
    """
    print(f"\n{role.capitalize()} Login")
    username = input("Enter your username: ").strip()
    password = input("Enter your password: ").strip()

    if username == credentials[role]["username"] and password == credentials[role]["password"]:
        print(f"Welcome, {role.capitalize()}!")
        return True
    else:
        print("Invalid credentials. Please try again.")
        return False

# AI Functionalities for User Dashboard
def ai_course_recommendation():
    """
    AI-based course recommendation system for users.
    """
    available_courses = [
        "Effective Communication Skills",
        "Time Management Mastery",
        "Conflict Resolution",
        "Leadership Essentials",
        "Emotional Intelligence Basics",
    ]
    print("\nBased on your interest, here are some recommended courses:")
    recommendations = random.sample(available_courses, 3)  # Randomly recommend 3 courses
    for idx, course in enumerate(recommendations, start=1):
        print(f"{idx}. {course}")
    choice = input("Select a course to enroll (1/2/3 or type 'skip'): ").strip()
    if choice in ["1", "2", "3"]:
        print(f"You have successfully enrolled in '{recommendations[int(choice)-1]}'. Happy learning!")
    elif choice.lower() == "skip":
        print("No problem! Explore the courses anytime.")
    else:
        print("Invalid input. Returning to User Dashboard.")

def ai_learning_support():
    """
    AI-powered chatbot to assist users with learning queries.
    """
    print("\nWelcome to the AI Learning Support Assistant!")
    print("Type your query or type 'exit' to return to the dashboard.")
    while True:
        query = input("Your query: ").strip()
        if query.lower() == "exit":
            print("Exiting AI Learning Support. Have a great day!")
            break
        else:
            responses = {
                "how to improve communication": "Improving communication requires active listening, clarity in speech, and empathy.",
                "time management tips": "Prioritize your tasks, avoid multitasking, and use tools like to-do lists or calendars.",
                "leadership skills": "Leadership requires building trust, inspiring others, and making informed decisions.",
            }
            print(responses.get(query.lower(), "I'm here to help! Could you rephrase your question?"))

# Voice-based Pronunciation Coaching using gTTS
def pronunciation_coaching():
    """
    Provides voice-based coaching to improve pronunciation (using gTTS).
    """
    print("\nPronunciation Coaching:")
    word = input("Enter a word you'd like to practice: ").strip()
    print(f"\nI will now pronounce the word '{word}' for you.")

    # Use gTTS to convert the word into speech and play it
    tts = gTTS(text=word, lang='en')
    tts.save("pronunciation.mp3")
    display(Audio("pronunciation.mp3"))

# User Dashboard
def user_dashboard():
    """
    User-specific functionalities with AI features.
    """
    while True:
        print("\nUser Dashboard:")
        print("Options:\n1. View assigned courses\n2. Take a quiz\n3. AI Course Recommendation\n4. AI Learning Support\n5. Pronunciation Coaching\n6. Exit")
        choice = input("Enter your choice (1/2/3/4/5/6): ").strip()

        if choice == "1":
            print("Viewing assigned courses...")
            print("Assigned Courses: [Course1, Course2, Course3]")
        elif choice == "2":
            print("Starting a quiz...")
            question = "What is the key to effective communication?"
            options = ["1. Listening", "2. Speaking", "3. Writing", "4. Reading"]
            print(f"Question: {question}")
            print("Options:", ", ".join(options))
            answer = input("Enter your answer (1/2/3/4): ").strip()
            if answer == "1":
                print("Correct! Listening is key to effective communication.")
            else:
                print("Incorrect. The correct answer is 'Listening'.")
        elif choice == "3":
            ai_course_recommendation()
        elif choice == "4":
            ai_learning_support()
        elif choice == "5":
            pronunciation_coaching()
        elif choice == "6":
            print("Exiting User Dashboard.")
            break
        else:
            print("Invalid choice. Please try again.")

# HR Dashboard
def hr_dashboard():
    """
    HR-specific functionalities.
    """
    while True:
        print("\nHR Dashboard:")
        print("Options:\n1. View course completion status\n2. Assign new courses\n3. View feedback\n4. Exit")
        choice = input("Enter your choice (1/2/3/4): ").strip()

        if choice == "1":
            print("Viewing course completion status...")
        elif choice == "2":
            print("Assigning new courses...")
        elif choice == "3":
            print("Viewing feedback...")
        elif choice == "4":
            print("Exiting HR Dashboard.")
            break
        else:
            print("Invalid choice. Please try again.")

# Admin Dashboard
def admin_dashboard():
    """
    Admin-specific functionalities.
    """
    while True:
        print("\nAdmin Dashboard:")
        print("Options:\n1. Manage Users\n2. Generate Reports\n3. Exit")
        choice = input("Enter your choice (1/2/3): ").strip()

        if choice == "1":
            print("Managing users...")
        elif choice == "2":
            print("Generating reports...")
        elif choice == "3":
            print("Exiting Admin Dashboard.")
            break
        else:
            print("Invalid choice. Please try again.")

# Login System
def login_system():
    """
    Function to handle user login and role selection.
    """
    print("Welcome to the Soft Skills Enhancement App!")
    print("Login as:\n1. Admin\n2. HR Manager\n3. User (Candidate)")
    role = input("Select your role (1/2/3): ").strip()

    if role == "1":
        if authenticate("admin"):
            admin_dashboard()
    elif role == "2":
        if authenticate("hr"):
            hr_dashboard()
    elif role == "3":
        if authenticate("user"):
            user_dashboard()
    else:
        print("Invalid selection. Please try again.")
        login_system()

# Main Function
def main():
    """
    Main function to run the application.
    """
    login_system()

# Run the application
if __name__ == "__main__":
    main()
```
