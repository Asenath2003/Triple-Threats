```python
import random
# Initialize user progress tracking
user_progress = {
   "Vocabulary": {"Learned": 0},
   "Speech Lessons": {"Completed": 0},
   "Pronunciation Lessons": {"Exercises Completed": 0},
   "Practice Tests": {"Tests Taken": 0, "Average Score": 0},
}
# Predefined user credentials for authentication
credentials = {
   "admin": {"username": "admin", "password": "admin123"},
   "hr": {"username": "hr", "password": "hr123"},
   "user": {"username": "user", "password": "user123"},
}
# Authentication Function
def authenticate(role):
   """
   Authenticates a user based on their role.
   """
   print(f"\n{role.capitalize()} Login")
   username = input("Enter your username: ").strip()
   password = input("Enter your password: ").strip()
   if username == credentials[role]["username"] and password == credentials[role]["password"]:
       print(f"Welcome, {role.capitalize()}!")
       return True
   else:
       print("Invalid credentials. Please try again.")
       return False
# Vocabulary Lessons
def vocabulary_lessons():
   print("\nVocabulary Lessons:")
   print("1. Interactive Modules (Quizzes/Flashcards)")
   print("2. Progress Tracking")
   choice = input("Select an option (1/2): ").strip()
   if choice == "1":
       print("\nInteractive Vocabulary Quiz:")
       questions = [
           {"question": "What is the synonym of 'Happy'?", "options": ["A. Sad", "B. Joyful", "C. Angry"], "answer": "B"},
           {"question": "What is the opposite of 'Cold'?", "options": ["A. Hot", "B. Warm", "C. Cool"], "answer": "A"},
       ]
       score = 0
       for q in questions:
           print(f"\nQuestion: {q['question']}")
           print("Options:", ", ".join(q["options"]))
           answer = input("Enter your answer (A/B/C): ").strip().upper()
           if answer == q["answer"]:
               print("Correct!")
               score += 1
           else:
               print(f"Incorrect. The correct answer was '{q['answer']}'.")
       user_progress["Vocabulary"]["Learned"] += score
       print(f"\nYou scored {score}/{len(questions)}. Total words learned: {user_progress['Vocabulary']['Learned']}.")
   elif choice == "2":
       print(f"Progress: Words learned so far: {user_progress['Vocabulary']['Learned']}.")
   else:
       print("Invalid choice. Returning to dashboard.")
# Speech Lessons
def speech_lessons():
   print("\nSpeech Lessons:")
   print("1. Audio Lessons")
   print("2. Interactive Dialogues")
   choice = input("Select an option (1/2): ").strip()
   if choice == "1":
       print("\nPlaying an audio lesson to improve speaking skills...")
       print("Lesson: 'How to confidently introduce yourself'.")
       user_progress["Speech Lessons"]["Completed"] += 1
       print(f"Lessons completed: {user_progress['Speech Lessons']['Completed']}.")
   elif choice == "2":
       print("\nSimulated Conversation: Practice introducing yourself.")
       user_input = input("Your turn: ").strip()
       print(f"Good job! You said: '{user_input}'. Keep practicing!")
   else:
       print("Invalid choice. Returning to dashboard.")
# Pronunciation Lessons
def pronunciation_lessons():
   print("\nPronunciation Lessons:")
   print("1. Phonetic Guides")
   print("2. Practice Exercises")
   choice = input("Select an option (1/2): ").strip()
   if choice == "1":
       print("\nPhonetic Guide: Learn to pronounce 'Entrepreneur' (on-truh-pruh-nur).")
   elif choice == "2":
       print("\nPractice Exercise: Pronounce 'Innovation'.")
       print("Feedback: Good job! Your pronunciation is accurate.")
       user_progress["Pronunciation Lessons"]["Exercises Completed"] += 1
       print(f"Exercises completed: {user_progress['Pronunciation Lessons']['Exercises Completed']}.")
   else:
       print("Invalid choice. Returning to dashboard.")
# Practice Tests
def practice_tests():
   print("\nPractice Tests:")
   print("1. Multiple Choice Test")
   print("2. Fill-in-the-Blank Test")
   choice = input("Select an option (1/2): ").strip()
   if choice == "1":
       print("\nMultiple Choice Test:")
       questions = [
           {"question": "What is the synonym of 'Begin'?", "options": ["A. Start", "B. End", "C. Pause"], "answer": "A"},
       ]
       score = 0
       for q in questions:
           print(f"\nQuestion: {q['question']}")
           print("Options:", ", ".join(q["options"]))
           answer = input("Enter your answer (A/B/C): ").strip().upper()
           if answer == q["answer"]:
               print("Correct!")
               score += 1
           else:
               print(f"Incorrect. The correct answer was '{q['answer']}'.")
       user_progress["Practice Tests"]["Tests Taken"] += 1
       user_progress["Practice Tests"]["Average Score"] = score
       print(f"Test completed. Your score: {score}/{len(questions)}.")
   elif choice == "2":
       print("\nFill-in-the-Blank Test:")
       print("Complete the sentence: 'I ____ to the store yesterday.'")
       answer = input("Your answer: ").strip()
       if answer.lower() == "went":
           print("Correct!")
       else:
           print("Incorrect. The correct answer is 'went'.")
       user_progress["Practice Tests"]["Tests Taken"] += 1
       print("Test completed.")
   else:
       print("Invalid choice. Returning to dashboard.")
# Test Reports
def test_reports():
   print("\nTest Reports:")
   print(f"Tests Taken: {user_progress['Practice Tests']['Tests Taken']}.")
   print(f"Average Score: {user_progress['Practice Tests']['Average Score']}.")
   print("Recommendations: Focus on vocabulary building for improved results.")
# Learning Path
def learning_path():
   print("\nLearning Path:")
   print("Based on your progress, here's a recommended path:")
   print("- Focus on Vocabulary Lessons for next week.")
   print("- Practice Speech Lessons for 10 minutes daily.")
   print("- Take a Pronunciation Exercise every alternate day.")
   print("Milestone: Improve test scores by 20% in the next assessment.")
# User Dashboard
def user_dashboard():
   while True:
       print("\nUser Dashboard:")
       print("Options:\n1. Vocabulary Lessons\n2. Speech Lessons\n3. Pronunciation Lessons\n4. Practice Tests\n5. Test Reports\n6. Learning Path\n7. Exit")
       choice = input("Enter your choice (1/2/3/4/5/6/7): ").strip()
       if choice == "1":
           vocabulary_lessons()
       elif choice == "2":
           speech_lessons()
       elif choice == "3":
           pronunciation_lessons()
       elif choice == "4":
           practice_tests()
       elif choice == "5":
           test_reports()
       elif choice == "6":
           learning_path()
       elif choice == "7":
           print("Exiting User Dashboard.")
           break
       else:
           print("Invalid choice. Please try again.")
# HR Dashboard
def hr_dashboard():
   """
   HR-specific functionalities.
   """
   while True:
       print("\nHR Dashboard:")
       print("Options:\n1. View course completion status\n2. Assign new courses\n3. View feedback\n4. Exit")
       choice = input("Enter your choice (1/2/3/4): ").strip()
       if choice == "1":
           print("Viewing course completion status...")
       elif choice == "2":
           print("Assigning new courses...")
       elif choice == "3":
           print("Viewing feedback...")
       elif choice == "4":
           print("Exiting HR Dashboard.")
           break
       else:
           print("Invalid choice. Please try again.")
# Admin Dashboard
def admin_dashboard():
   """
   Admin-specific functionalities.
   """
   while True:
       print("\nAdmin Dashboard:")
       print("Options:\n1. Manage Users\n2. Generate Reports\n3. Exit")
       choice = input("Enter your choice (1/2/3): ").strip()
       if choice == "1":
           print("Managing users...")
       elif choice == "2":
           print("Generating reports...")
       elif choice == "3":
           print("Exiting Admin Dashboard.")
           break
       else:
           print("Invalid choice. Please try again.")
# Main Function
def main():
   """
   Main function to handle user login and navigation.
   """
   print("Welcome to the Soft Skills Enhancement App!")
   print("Login as:\n1. Admin\n2. HR Manager\n3. User (Candidate)")
   role = input("Select your role (1/2/3): ").strip()
   if role == "1":
       if authenticate("admin"):
           admin_dashboard()
   elif role == "2":
       if authenticate("hr"):
           hr_dashboard()
   elif role == "3":
       if authenticate("user"):
           user_dashboard()
   else:
       print("Invalid selection. Please try again.")
# Run the main function
if __name__ == "__main__":
  main()
```
